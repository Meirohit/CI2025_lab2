```python
import numpy as np
from itertools import combinations
import os

# Get all .npy files in the lab2 directory
lab2_path = 'lab2'
npy_files = sorted([f for f in os.listdir(lab2_path) if f.endswith('.npy') and f.startswith('problem_')])

# Iterate through each file
for filename in npy_files:
    print(f"\n{'='*60}")
    print(f"File: {filename}")
    print('='*60)
    
    # Load the problem
    problem = np.load(os.path.join(lab2_path, filename))
    num_cities = problem.shape[0]
    
    print(f"Number of cities: {num_cities}")
    print(f"Has negative values: {np.any(problem < 0)}")
    print(f"Diagonal is all zero: {np.allclose(np.diag(problem), 0.0)}")
    print(f"Is symmetric: {np.allclose(problem, problem.T)}")
    print(f"Satisfies triangular inequality: {all(problem[x, y] <= problem[x, z] + problem[z, y] for x, y, z in combinations(range(problem.shape[0]), 3))}")
```
